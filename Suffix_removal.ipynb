```python
import pandas as pd
import sklearn as sk
import math 
```

```python
with open("Sri Lanka Constitution-Sinhala.txt", "r") as file:
    text = file.read()
```

```python
with open('suffixes_list.txt', 'r') as file:
    # Read lines and strip newline characters
    suffixes = [line.strip() for line in file]
```

```python
suffixes
```

```python
len(suffixes)
```

```python
text
```

```python
text = text.split(" ")
```

```python
text
```

```python
new_text = []
for word in text:
  for suffix in suffixes:
    if word.lower().endswith(suffix.lower()):
      word = word[0:-len(suffix)]
  new_text.append(word)
```

```python
new_text
```

```python
len(list(set(new_text)))
```

```python
len(list(set(text)))
```

```python
concatenated_string = ' '.join(new_text)
```

```python
concatenated_string
```

```python
with open('Sri Lanka Constitution-Sinhala_Suff_Rem.txt', 'w') as file:
    file.write(concatenated_string)
```
