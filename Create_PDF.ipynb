```python
import PyPDF2
```

```python
page_content = []
with open('Sri Lanka Constitution - Sinhala.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        text = page.extract_text()
        page_content.append(text)
```

```python
with open('suffixes_list.txt', 'r') as file:
    # Read lines and strip newline characters
    suffixes = [line.strip() for line in file]
```

```python
processed_pages = []
for i in range(len(page_content)):
  text = page_content[i].split(" ")
  new_text = []
  concatenated_string = ' '
  for word in text:
    for suffix in suffixes:
      if word.lower().endswith(suffix.lower()):
        word = word[0:-len(suffix)]
    new_text.append(word)
  concatenated_string = ' '.join(new_text)
  processed_pages.append(concatenated_string)
```

```python
for i in range(len(processed_pages)):
    with open(f'Text_files/Sri Lanka Constitution-Sinhala_Suff_Rem_Page_{i+1}.txt', 'w') as file:
        file.write(processed_pages[i])
```
